```python
# Define Chemical System Parameters
k4, k2, k1  = (0.00140174869155172, 0.010761125573375087, 0.005158278889342589)
system_name = 'Adenine-Thymine'
savedir     = 'Results'

AT_data = Chemical_KerrCat_Analysis(N_basis=N, system_name="Adenine_Thymine",
                                    mass=mass, k1=k1, k2=k2, k4=k4, savedir='Results')
AT_data.initialize_plotting_data(xplot_min=-4, xplot_max=4, N_xplot=N)
AT_data.initialize_cvalues(c_min=delta_c, c_max=0.6+delta_c, c_step=delta_c, do_phi_ZPS=False)
AT_data.loop_over_c(N_check=40)
AT_data.plot_energies_and_parameters(n_plotted=12, plot_interval=3, save_figs=True)
```
